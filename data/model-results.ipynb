```python
import numpy as np
from matplotlib import pyplot as plt

#plt.rcParams["figure.figsize"] = (8,5)
```

```python
distracted = {
    "Naive Bayes": [0.828, 0.821, 0.809],
    "Decision Table": [0.881, 0.888, 0.905],
    "Random Forest": [0.896, 0.895, 0.905]
}
inattentive = {
    "Naive Bayes": [0.809, 0.809, 0.756],
    "Decision Table": [0.794, 0.777, 0.869],
    "Random Forest": [0.833, 0.833, 0.833]
}
distracted_inattentive = {
    "Naive Bayes": [0.815, 0.833, 0.737],
    "Decision Table": [0.865, 0.833, 0.821],
    "Random Forest": [0.881, 0.889, 0.854]
}
```

```python
windows = ["10s", "20s", "30s"]

plt.plot(windows, distracted["Naive Bayes"], marker = 'o', label = "Naive Bayes (NB)")
plt.plot(windows, distracted["Decision Table"], marker = '*', label = "Decision Table (DT)")
plt.plot(windows, distracted["Random Forest"], marker = 's', label = "Random Forest (RF)")
axes = plt.gca()
axes.set_ylim([0.5, 1.0])
plt.title("Distracted detection models")
plt.xlabel("Window size (seconds)")
plt.ylabel("F1-score")
plt.legend(loc = 'lower left')
plt.show()
```

```python
windows = ["10s", "20s", "30s"]

plt.plot(windows, inattentive["Naive Bayes"], marker = 'o', label = "Naive Bayes (NB)")
plt.plot(windows, inattentive["Decision Table"], marker = '*', label = "Decision Table (DT)")
plt.plot(windows, inattentive["Random Forest"], marker = 's', label = "Random Forest (RF)")
axes = plt.gca()
axes.set_ylim([0.5, 1.0])
plt.title("Inattentive detection models")
plt.xlabel("Window size (seconds)")
plt.ylabel("F1-score")
plt.legend(loc = 'lower left')
plt.show()
```

```python
windows = ["10s", "20s", "30s"]

plt.plot(windows, distracted_inattentive["Naive Bayes"], marker = 'o', label = "Naive Bayes (NB)")
plt.plot(windows, distracted_inattentive["Decision Table"], marker = '*', label = "Decision Table (DT)")
plt.plot(windows, distracted_inattentive["Random Forest"], marker = 's', label = "Random Forest (RF)")
axes = plt.gca()
axes.set_ylim([0.5, 1.0])
plt.title("Distracted/inattentive detection models")
plt.xlabel("Window size (seconds)")
plt.ylabel("F1-score")
plt.legend(loc = 'lower left')
plt.show()
```
